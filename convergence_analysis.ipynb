```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
plt.rcParams['figure.dpi'] = 200
from scipy.interpolate import make_interp_spline 

delta = 15
window_siz = 10000
```

```python
# This function rolls a sliding window of length 1000 and sees what percent of those 1000 time steps are stable
def get_values_to_plot(arr, window_size):
    returns = [0] 
    for index in range(len(arr) - window_size): 
        curr_window = arr[index: index + window_size].copy()
        count = 0
        for item in curr_window: 
            if item >= 100-delta:
                count += 1 
        returns.append(count/window_size) 
    return returns

# This function reads in the CSV file, extracts the stability column and returns it as an array 
def read_in_dataframe(file_location):
    df = pd.read_csv(file_location)
    df = df[['Average Stability']][1:]
    arr = df.to_numpy()[:,0]
    return arr
```

```python
colors1 = [ '#000000', '#ff8d03', '#ffb303',  '#ffd503']
colors2 = ['#000000', '#03ffc0', '#03fff7', '#03cdff']

def write_output(filename, array):
    np.savetxt(filename, array) 


def read_in_txt(filename): 
    df = pd.read_csv(filename, sep=',', header=None).to_numpy()
    arr = []
    for item in df: 
        arr.append(item[0])
    arr = np.array(arr)
    return arr 

loc0 = r"/Users/gaurab/Documents/Git/logs/analysis/"
loc1 = r"/Users/gaurab/Documents/Git/aalgs/temp/"
loc = r"/Users/gaurab/Documents/Git/aalgs/temp/"
```

```python
location = loc0 + "UCB/rnd rnd"
plot_lists_ucb_rr = []
legend_ucb_rr = []
vars = [5, 10, 15, 20]
# MAKE LISTS TO PLOT RANDOM RANDOM
for var in vars: 
    print(f"Market size {var}")
    filename = location + f"/MASTER_N_{var}_stability.csv"
    lst = get_values_to_plot(read_in_dataframe(filename), window_siz)
    plot_lists_ucb_rr.append(lst)
    write_output(loc1 + f"N_{var}_ucb_rr_delta_{delta}.txt", lst)
    legend_ucb_rr.append(f"N = {var}")
```
